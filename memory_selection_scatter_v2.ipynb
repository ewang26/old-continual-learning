```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p_arr = [0.001, 0.002, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1]
mem_arr = ['random', 'GSS']
dataset_name = 'MNIST_split'
data_path = f'similarity_metric_plots/{dataset_name}'
```

```python
for p_index, p in enumerate(p_arr):
    plt.figure()
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.xlabel('accuracy')
    plt.ylabel('cosine similarity')
    plt.title(f'Scatter plot of cosine similarity vs train acc for p = {p}')
    for data_index, mem in enumerate(mem_arr):
        acc_arr = np.load(f'{data_path}/{mem}/acc_cos_sim.npy')
        sim_arr = np.load(f'{data_path}/{mem}/past_cos_sim.npy')
        plt.scatter(acc_arr[p_index], sim_arr[p_index], label = f'{mem}')
    plt.legend()
    plt.savefig(f'memory_selection_plots/{dataset_name}/compare_p_{p}.png')
    plt.show()
    plt.clf()
```
