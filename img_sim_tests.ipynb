```python
import numpy as np
import matplotlib.pyplot as plt
from image_similarity_measures.quality_metrics import rmse, issm, psnr, ssim, uiq
```

```python
def vec_to_img(vec):
    '''
    vec = 1d numpy array
    '''
    img_length = int(len(vec)**0.5)
    return np.reshape(vec, (img_length, img_length))[..., np.newaxis]
```

```python
v1 = np.load(f'models/mnist_split/1/train/task_1/grad_layers.0.weight.npy')[2]
v2 = np.load(f'models/mnist_split/0.001/train/task_1/grad_layers.0.weight.npy')[2]
i1 = vec_to_img(v1)
i2 = vec_to_img(v2)
plt.imshow(i1)
plt.show()
plt.imshow(i2)
plt.show()
```

```python
def get_rmse(i1, i2):
    max_p = max(np.max(i1), np.max(i2))
    return rmse(i1, i2, max_p)

def get_issm(i1, i2):
    return issm(i1, i2)

def get_psnr(i1, i2):
    max_p = max(np.max(i1), np.max(i2))
    return psnr(i1, i2, max_p)

def get_ssim(i1, i2):
    max_p = max(np.max(i1), np.max(i2))
    return ssim(i1, i2, max_p)

def get_uiq(i1, i2):
    return uiq(i1, i2)
```

```python

for func in [get_rmse, get_issm, get_psnr, get_ssim, get_uiq]:
    print(func(i1, i2))
```
